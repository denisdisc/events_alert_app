```python
import requests
import json
```

```python
def get_payload(url):
    url = f'https://afisha.ru/{url}'
    with requests.post(url=url, headers={"Accept": 'application/json'}) as response:
        return json.loads(response.text)
```

```python
movie_json = get_payload('movie/movies_list')['Movies']['Items']
```

```python
concert_json = get_payload('msk/concerts')['Widget']['Tiles']
```
